```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpages =  requests.get("https://www.ambitionbox.com/list-of-companies?page=1", headers=headers).text
```

```python
soup = BeautifulSoup(webpages)
```

```python
company = soup.find_all('div',class_="company-content-wrapper")
```

```python
len(company)
```

```python
name = []
rating = []
review = []
ComType = []
hq = []
old = []
empl = []
```

```python

for i in company:
     name.append( i.find('h2').text.strip())
     rating.append( i.find('p',class_='rating').text.strip())
     review.append( i.find('a',class_= 'review-count').text.strip())
     ComType.append( i.find('p',class_='infoEntity').text.strip())
     hq.append( i.find_all('p',class_='infoEntity')[1].text.strip())
     old.append( i.find_all('p',class_='infoEntity')[2].text.strip())
     empl.append( i.find_all('p',class_='infoEntity')[3].text.strip())
     
     
     
```

```python
d = { 'name': name, 'rating': rating , 'review': review , 'Type': ComType , 'Address': hq, 'old': old , 'employee': empl }

df = pd.DataFrame(d)
df
```
